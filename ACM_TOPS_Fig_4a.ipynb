```python
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files

# Initialize
p1 = np.arange(0, 1.01, 0.01)
p2 = np.arange(0, 1.01, 0.01)
p0 = 0.5
P1, P2 = np.meshgrid(p1, p2)
delta = 0.05
q1 = np.arange(max(min(p1) - delta, 0), min(max(p1) + delta, 1) + 0.01, 0.01)
q2 = np.arange(max(min(p2) - delta, 0), min(max(p2) + delta, 1) + 0.01, 0.01)
Q1, Q2 = np.meshgrid(q1, q2)

# Green sub-region
I1 = (p0 * P1) < (1 - p0) * (1 - P2) - delta
I2 = P2 * (1 - p0) < p0 * (1 - P1) - delta
I3 = np.logical_and(I1, I2)
plt.scatter(P1[I3], P2[I3], c='#00FF00')

# Magenta sub-region
I1 = (P1 * p0) < (1 - P2) * (1 - p0)
I2 = P2 * (1 - p0) < p0 * (1 - P1)
I3 = p0 * P1 > (1 - P2) * (1 - p0) - delta
I4 = P2 * (1 - p0) > p0 * (1 - P1) - delta
I5 = np.logical_and(I1, I2)
I6 = np.logical_and(I3, I4)
I7 = np.logical_and(I5, I6)
plt.scatter(P1[I7], P2[I7], c='#FF00FF')

# Red sub-region
I1 = (P1 * p0) > (1 - P2) * (1 - p0)
I2 = P2 * (1 - p0) > p0 * (1 - P1)
I3 = p0 * P1 < (1 - P2) * (1 - p0) + delta
I4 = P2 * (1 - p0) < p0 * (1 - P1) + delta
I5 = np.logical_and(I1, I2)
I6 = np.logical_and(I3, I4)
I7 = np.logical_and(I5, I6)
plt.scatter(P1[I7], P2[I7], c='#FF0000')

# Cyan sub-region
I1 = (p0 * P1) > (1 - p0) * (1 - P2) + delta
I2 = P2 * (1 - p0) > p0 * (1 - P1) + delta
I3 = np.logical_and(I1, I2)
plt.scatter(P1[I3], P2[I3], c='#00FFFF')

plt.gca().spines['top'].set_visible(False)
plt.gca().spines['bottom'].set_visible(False)
plt.gca().spines['left'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

plt.xlabel('$p_1, q_1 \\rightarrow$', fontsize = 18)
plt.ylabel('$p_2, q_2 \\rightarrow$', fontsize = 18)
plt.title('Regions for $(p_1, p_2)$ and $(q_1, q_2)$ when $p_0$ = 0.5', fontsize = 16)

plt.savefig('region4_new.jpg')
files.download('region4_new.jpg')


plt.show()
```
